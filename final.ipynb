f(x) = sin(x)^2 - cos(x)^2

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

```python
import sympy
```

# Найдем корни уравнения

```python
expr = 'sin(x)**2 - cos(x)**2'

def solve_eq(expr: str) -> str:               
    x = sympy.Symbol('x')
    solution = sympy.solveset(expr, x, domain=sympy.S.Reals)
    return solution

solve_eq(expr)
```

# Найдем производную

```python
def derivative(expr: str):
    x = sympy.Symbol('x')
    derivative = sympy.diff(expr)
    return derivative

derivative(expr)
```

# Интервалы где функция возрастает

```python
def increases(expr: str):
    x = sympy.Symbol('x')
    increase = sympy.solve(expr>0)
    return increase
increases(derivative(expr))
```

# Интервалы где функция убывает

```python
def decreases(expr: str):
    x = sympy.Symbol('x')    
    decrease = sympy.solve(expr<0)
    return decrease

decreases(derivative(expr))
```

# График функции

```python
def plot(expr: str):
    x = sympy.Symbol('x')
    plot = sympy.plot(expr)
    return plot
plot(expr)
```

# Вершины

```python
solve_eq(derivative(expr))
```

# Промежутки, на котором f > 0 

```python
expr = 'sin(x)**2 - cos(x)**2>0'
x = sympy.Symbol('x')
sympy.solve(expr)
```

# Промежутки, на котором f < 0

```python
expr = 'sin(x)**2 - cos(x)**2<0'
x = sympy.Symbol('x')
sympy.solve(expr)
```
